```python
def draw_mark(image, results):
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
```

### ce code extrait les feutures:
- on calcule la distance entre le point de jonction 0 de la main et l'apaule
- on calcule la distance entre chaque points des doights et le point de jonction 0 de la main, ceci nous permet d'avoir la forme de la main a l'instant t.
- on divise chque vector par le max pour normaliser 

ansi nous avons un point de refernce qui bouge avec le corp donc la variation des coordonées selon la profondeur est élimnée.




```python
f_tip=np.zeros((6,3))
f_tip2=np.zeros((6,3))
f_tip_list = [0,4,8,12,16,20]  #number of the points in the landmark choisi

f_tip[0] = abs(hand_mark[0]-pose[12][:-1])
f_tip2[0] = abs(hand_mark2[0]-pose2[12][:-1])

f_tip[0]=f_tip[0]/max(f_tip[0])
f_tip2[0]=f_tip2[0]/max(f_tip2[0])

for i in range(1,6):
    for j in range(3):
        f_tip[i][j] = abs(hand_mark[f_tip_list[i]][j]-hand_mark[f_tip_list[0]][j])
        f_tip2[i][j] = abs(hand_mark2[f_tip_list[i]][j]-hand_mark2[f_tip_list[0]][j])
   
    f_tip[i] = f_tip[i]/max(f_tip[i])
    f_tip2[i] = f_tip2[i]/max(f_tip2[i])

print(f_tip)
print("\n")
print(f_tip2)
print("\n")
```
